# Criação de variaveis de player_totals

* Pontos
* Assistências
* Bloqueio
* Roubos
* Jogos
* Minutos

Link de referencia para nomes das colunas https://www.basketball-reference.com/about/glossary.html

```python
# Reatribuindo dataframe para mais simples vizualização
hof=player_career_info

# Lista de jogadores
list=hof['player']

pontos=[]
assistencias=[]
rebotes=[]
bloqueios=[]
roubos=[]
jogos=[]
minutos=[]

for i in list:
    pontos.append(player_totals[player_totals['player']==i].sum()['pts'])
    assistencias.append(player_totals[player_totals['player']==i].sum()['ast'])
    rebotes.append(player_totals[player_totals['player']==i].sum()['trb'])
    bloqueios.append(player_totals[player_totals['player']==i].sum()['blk'])
    roubos.append(player_totals[player_totals['player']==i].sum()['stl'])
    jogos.append(player_totals[player_totals['player']==i].sum()['g'])
    minutos.append(player_totals[player_totals['player']==i].sum()['mp'])

# adição de variáveis no dataframe principal
hof['pontos']=pontos
hof['assistencias']=assistencias    
hof['rebotes']=rebotes
hof['bloqueios']=bloqueios
hof['roubos']=roubos
hof['jogos']=jogos
hof['minutos']=minutos
```

```python
allnba1=[]
for i in list:
    a=end_of_season[(end_of_season['type']=='All-NBA')&(end_of_season['player']==i)]
    if a.empty:
        x=0     
    elif a['number_tm'].value_counts().count()==3:
        x=a['number_tm'].value_counts()['1st']
    elif a['number_tm'].value_counts().count()==2:
        if (a['number_tm'].value_counts().index[0]=='1st') | (a['number_tm'].value_counts().index[1]=='1st'):
            x=a['number_tm'].value_counts()['1st']
        else:
            x=0
    else:
        if a['number_tm'].value_counts().index[0]!='1st':
            x=0
        else:
            x=a['number_tm'].value_counts()['1st']
            
    allnba1.append(x)    
hof['All_NBA_1st']=allnba1   
```

```python
allnba2=[]
for i in list:
    a=end_of_season[(end_of_season['type']=='All-NBA')&(end_of_season['player']==i)]
    if a.empty:
        x=0     
    elif a['number_tm'].value_counts().count()==3:
        x=a['number_tm'].value_counts()['2nd']
    elif a['number_tm'].value_counts().count()==2:
        if (a['number_tm'].value_counts().index[0]=='2nd') | (a['number_tm'].value_counts().index[1]=='2nd'):
            x=a['number_tm'].value_counts()['2nd']
        else:
            x=0
    else:
        if a['number_tm'].value_counts().index[0]!='2nd':
            x=0
        else:
            x=a['number_tm'].value_counts()['2nd']
            
    allnba2.append(x)    
hof['All_NBA_2nd']=allnba2   
```

```python
allnba3=[]
for i in list:
    a=end_of_season[(end_of_season['type']=='All-NBA')&(end_of_season['player']==i)]
    if a.empty:
        x=0     
    elif a['number_tm'].value_counts().count()==3:
        x=a['number_tm'].value_counts()['3rd']
    elif a['number_tm'].value_counts().count()==2:
        if (a['number_tm'].value_counts().index[0]=='3rd') | (a['number_tm'].value_counts().index[1]=='3rd'):
            x=a['number_tm'].value_counts()['3rd']
        else:
            x=0
    else:
        if a['number_tm'].value_counts().index[0]!='3rd':
            x=0
        else:
            x=a['number_tm'].value_counts()['3rd']
            
    allnba3.append(x)    
hof['All_NBA_3rd']=allnba3        
```

```python
allnbadef=[]
for i in list:
    a=end_of_season[(end_of_season['type']=='All-Defense')&(end_of_season['player']==i)]
    if a.empty:
        x=0     
    elif a['number_tm'].value_counts().count()==2:
        x=a['number_tm'].value_counts()['1st']
    else:
        if a['number_tm'].value_counts().index[0]!='1st':
            x=0
        else:
            x=a['number_tm'].value_counts()['1st']
            
    allnbadef.append(x)    
hof['All_NBA_Def_1st']=allnbadef        
```

```python
allstar=[]
for i in list:
    a=all_star[all_star['player']==i]
    if a.empty:
        x=0     
    else:
        x=a['replaced'].count()
            
    allstar.append(x)    
hof['Marcacao_All_Star']=allstar        
```

```python
mvp=[]
for i in list:
    a=player_award_shares[(player_award_shares['award']=='nba mvp')&(player_award_shares['player']==i)]
    if a.empty:
        x=0     
    else:
        x=sum(a['winner'])
            
    mvp.append(x)    
hof['MVPs']=mvp        
```
